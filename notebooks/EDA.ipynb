# Change ordinal features to int

```python
time_dict = {'7AM': 0, '10AM': 1, '2PM': 2, '6PM': 3, '10PM': 4}
expiration_dict = {'2h': 0, '1d': 1}
age_dict = {'below21': 0, '21': 1, '26': 2, '31': 3, '36': 4, '41': 5, '46': 6, '50plus': 7}
education_dict = {'Some High School': 0, 'High School Graduate': 1, 'Associates degree': 2,
             'Some college - no degree': 3, 'Bachelors degree': 4, 'Graduate degree (Masters or Doctorate)': 5}
income_dict = {'Less than $12500': 0, '$12500 - $24999': 1, '$25000 - $37499': 2, '$37500 - $49999': 3,
          '$50000 - $62499': 4, '$62500 - $74999': 5,'$75000 - $87499': 6, '$87500 - $99999': 7,
          '$100000 or More': 8}
bars_dict = {'never': 0, 'less1': 1, r'1~3': 2, r'4~8': 3, 'gt8': 4}
ordinal_dict = {'time': time_dict,
                'expiration': expiration_dict,
                'age': age_dict,
                'education': education_dict,
                'income': income_dict,
                'Bar': bars_dict,
                'CoffeeHouse': bars_dict,
                'CarryAway': bars_dict,
                'RestaurantLessThan20': bars_dict,
                'Restaurant20To50': bars_dict}


categorical_variables = ['time', 'expiration', 'age', 'education', 'income', 'Bar', 'CoffeeHouse', 
                                    'CarryAway', 'RestaurantLessThan20', 'Restaurant20To50']

def make_ordinal(df, categorical_variables):
    def make_ordinal_row(row, variable):
        row[variable] = ordinal_dict[variable][row[variable]]
        return row

    for variable in categorical_variables:
        df = df.apply(lambda x: make_ordinal_row(x, variable), axis=1)

    return df

df = make_ordinal(df, categorical_variables)

df.head(10)
```

# Change not ordinal features to one_hot

```python
# - destination (3 class)
# - passenger (4 class)
# - weather (3 class)
# - coupon (5 class) : hybrid
# - gender (2 class)
# - maritalStatus (5 class)
# - occupation (25 class)
# - car (5 class)


categorical_variables = ['destination', 'passanger', 'weather', 'coupon', 'gender', 'maritalStatus', 
                         'occupation'] 
for variable in categorical_variables:
    discarded = pd.get_dummies(df[variable],prefix=variable)
    if discarded.shape[1] == 2:
        discarded = discarded.iloc[:, 0]
    df= pd.concat([df, discarded], axis = 1)
    df.drop([variable], axis = 1, inplace = True)

df.head(10)
```

There are 0 records which are coupons for the same and opposite direction, which is ok, because these are excluding.

There are 0 records which neighter coupons for the same or opposite direction, which is ok, as these are completive. Therefore these are redundant informations and one of them can be dropped (direction_opp).

# Change the dataset according to the continual features

```python
df.drop(columns='direction_opp', inplace=True)

def label_coupon_time(row):
    if not row['toCoupon_GEQ5min']:
        return 0
    if row['toCoupon_GEQ5min'] and not row['toCoupon_GEQ15min']:
        return 1
    if row['toCoupon_GEQ15min'] and not row['toCoupon_GEQ25min']:
        return 2
    if row['toCoupon_GEQ25min']:
        return 3
    else:
        return None

df['coupon_time'] = df.apply(label_coupon_time, axis=1)
df.drop(columns='toCoupon_GEQ5min', inplace=True)
df.drop(columns='toCoupon_GEQ15min', inplace=True)
df.drop(columns='toCoupon_GEQ25min', inplace=True)
df.head(100)
```

```python
# feature importance
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

outputs = df.Y.to_numpy()
inputs_pd = df.drop(columns=['Y', 'Unnamed: 0'])
inputs = inputs_pd.to_numpy()

model = RandomForestClassifier(n_estimators=200, max_depth=5)
model.fit(inputs, outputs)

importance = model.feature_importances_
for i,(c, v) in enumerate(zip(inputs_pd.columns, importance)):
    print(f'Feature: {i} ({c}), Score: {v:.5f}')
# plot feature importance
plt.figure(figsize=(20, 10))
bars = plt.bar(inputs_pd.columns, importance)
plt.title('feature importances')
plt.xticks(rotation=90)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .002, f'{yval*100:.1f}%')
plt.show()
```
